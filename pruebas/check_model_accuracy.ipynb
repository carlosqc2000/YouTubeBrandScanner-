```python
import pandas as pd
import pymongo
# Conectar a MongoDB
def get_data_from_mongo():
    client = pymongo.MongoClient("mongodb://localhost:27017/")
    db = client["youtube_sponsors"]
    collection = db["sponsored_videos"]
    
    videos = collection.find({}, {"_id": 0, "sponsors": 1, "description": 1})
    
    data = []
    for video in videos:
        description = video.get("description", "No description")
        sponsors = ", ".join([sponsor["brand_name"] for sponsor in video.get("sponsors", [])])
        data.append({"Descripci√≥n": description, "Sponsors": sponsors})
    
    return pd.DataFrame(data)

# Exportar a Excel
def export_to_excel(df, filename="revision_sponsors2.xlsx"):
    df.to_excel(filename, index=False)
    print(f"‚úÖ Archivo exportado: {filename}")

# Ejecutar en un Notebook
if __name__ == "__main__":
    df = get_data_from_mongo()
    display(df)  # Mostrar en el notebook
    export_to_excel(df)
```
